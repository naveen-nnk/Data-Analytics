```python
dataframeMissingdata = pd.DataFrame({'Cricket':[1,2,np.nan,4,6,7,2,np.nan],
                  'Baseball':[5,np.nan,np.nan,5,7,2,4,5],
                  'Tennis':[1,2,3,4,5,6,7,8], 'Gender':[0,1,0,1,1,0,0,1]})
```

```python
dataframeMissingdata
```

```python
dataframeMissingdata.dropna()
```

```python
dataframeMissingdata.dropna(axis=1)       # Use axis=1 for dropping columns with nan values
```

```python
dataframeMissingdata.dropna(thresh=3)   #Thresh- minimum non null values
```

```python
dataframeMissingdata.fillna(value=0)
```

```python
dataframeMissingdata['Baseball'].mean()
```

```python
dataframeMissingdata['Baseball'].fillna(value=dataframeMissingdata['Baseball'].mean())
```
